# Notebook 01: Data Exploration
Run this notebook to explore the raw JSONL QA dataset.

```python
import json
import pandas as pd

def load_data(path):
    data = []
    with open(path, 'r') as f:
        for line in f:
            data.append(json.loads(line))
    return data

try:
    train = load_data('../data/processed/train_set.jsonl')
    print(f"Loaded {len(train)} training examples.")
    print(json.dumps(train[0], indent=2))
except FileNotFoundError:
    print("Run prepare_dataset.py first to generate the JSONL files.")
```
